```python
# concatenate 
df1 = pd.read_csv('data/dcm_2020_1.csv')
df2 = pd.read_csv('data/dcm_2020_2.csv')
df3 = pd.read_csv('data/dcm_2020_3.csv')
df4 = pd.read_csv('data/dcm_2020_4.csv')
df5 = pd.read_csv('data/dcm_2020_5.csv')
df6 = pd.read_csv('data/dcm_2020_6.csv')

df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
print(df.shape)
df.head()
```

```python
# what columns are there?
df.columns
```

```python
# remove unnecessary columns
df.drop(['Unnamed: 0', 'AmendedCharge', 'AmendedCode', 'AmendedCaseType', 'person_id', 'Complainant' ], axis=1, inplace=True)
print(df.shape)
```

```python
df.Locality.value_counts().nlargest(10)
```

```python
df.DefenseAttorney.value_counts().nlargest(10)
```

```python
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace('PD', 'PUBLIC DEFENDER')
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace('PUB DEF', 'PUBLIC DEFENDER')
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace('P DEF', 'PUBLIC DEFENDER')
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace('PUBLIC DEFENDER(TSR)', 'PUBLIC DEFENDER')

df.DefenseAttorney.value_counts().nlargest(10)
```

```python
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace("(", '')
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace(")", '')
df.DefenseAttorney.value_counts().nlargest(10)
```

```python
df['DefenseAttorney'] = df['DefenseAttorney'].str.replace('PUBLIC DEFENDERTSR', 'PUBLIC DEFENDER')
df.DefenseAttorney.value_counts().nlargest(10)
```

```python
# create empty column (initially fill it with nan values)
df['Attorney'] = np.nan

for i in range(len(df)):
    if type(df['DefenseAttorney'][i]) == float:
        df['Attorney'][i] = np.nan
    else:
        if df['DefenseAttorney'][i] == 'PUBLIC DEFENDER':
            df['Attorney'][i] = 'Public'
        else:
            df['Attorney'][i] = 'Private'
            
df.head()
```

```python
df.Attorney.value_counts().plot(kind='barh')

# drop 'DefenseAttorney' column
df.drop(['DefenseAttorney'], axis=1, inplace=True)
print(df.shape)
```

```python
# get only the zip codes from the address
df['zip'] = df['Address'].str.extract(r'(\d{5}\-?\d{0,4})')

# then drop Address column since we now have the column for zip codes
df.drop(['Address'], axis=1, inplace=True) 

df['zip'].head(10)
```

```python
df['Attorney'].isna().sum() / len(df)
```

```python
# drop columns with more than 45% of nan values
df = df.dropna(thresh=len(df) * 0.45, axis=1)
df.columns
```

```python
df.to_csv('data/dcm_2020.csv')
```

```python
df = pd.read_csv('data/dcm_2020.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True) 
df.head()
```

```python
# create 'duration' column (HearingDate - FiledDate)

# change data types
df['HearingDate'] = pd.to_datetime(df.HearingDate)
df['FiledDate'] = pd.to_datetime(df.FiledDate)


df['Duration'] = (df['HearingDate'] - df['FiledDate']).dt.days
df['Duration']
```

```python
df = pd.read_csv('data/dcm_2020.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True) 
df.head()
```

```python
options = ['Guilty', 'Guilty in Absentia', 'Nolle Prosequi', 'Dismissed', 'Not Guilty', 'Dismissed/Other', 'Not Guilty/Insanity']
fd = df[df['FinalDisposition'].isin(options)]
fd.head()
```

```python
# simplify final disposition to binary variable (either guilty or not guilty)
# create new column, 'fd'
df['fd'] = np.nan

guilty = ['Guilty', 'Guilty in Absentia']
no_guilty = ['Nolle Prosequi', 'Dismissed', 'Not Guilty', 'Dismissed/Other', 'Not Guilty/Insanity']

for i in range(len(df)):
    if df['FinalDisposition'][i] in guilty:
        df['fd'][i] = "guilty"
    elif df['FinalDisposition'][i] in no_guilty:
        df['fd'][i] = "no_guilty"
        
df['fd'].head()   
```

```python
df['fd'].value_counts().plot(kind='barh', color='red')
print(df['fd'].value_counts())
```

```python
df.to_csv('data/dcm_2020.csv')
```
